```python
from datetime import datetime

def calcular_dias_ate_hoje():
    try:
        data_str = input("Digite uma data no formato AAAA-MM-DD (ex: 2023-01-15): ")
        data_informada = datetime.strptime(data_str, '%Y-%m-%d')

        data_hoje = datetime.now()

        diferenca = data_hoje - data_informada

        print(f"Desde {data_informada.strftime('%Y-%m-%d')} até hoje ({data_hoje.strftime('%Y-%m-%d')}) se passaram {diferenca.days} dias.")

    except ValueError:
        print("Formato de data inválido. Por favor, use o formato AAAA-MM-DD.")
    except Exception as e:
        print(f"Ocorreu um erro: {e}")

calcular_dias_ate_hoje()
```

```python
# @title
import random
import statistics

def lancar_dado():
  """Simula o lançamento de um dado de 6 faces."""
  resultado = random.randint(1, 6)
  print(f"Você lançou o dado e o resultado foi: {resultado}")

def gerar_e_analisar_numeros():
  """Gera 5 números aleatórios e calcula sua média."""
  print("Gerando 5 números aleatórios...")
  lista_numeros = [random.randint(1, 100) for _ in range(5)] # Gera números entre 1 e 100
  print(f"Números gerados: {lista_numeros}")
  if lista_numeros:
    media = statistics.mean(lista_numeros)
    print(f"A média dos números gerados é: {media:.2f}")
  else:
    print("Não foi possível calcular a média de uma lista vazia.")

def verificar_numero_completo():
  """Verifica se um número é positivo/negativo/zero e par/ímpar."""
  try:
    num = float(input("Digite um número para verificar: "))

    # Positivo, Negativo ou Zero
    if num > 0:
      print(f"O número {num} é Positivo.")
    elif num < 0:
      print(f"O número {num} é Negativo.")
    else:
      print(f"O número {num} é Zero.")

    # Par ou Ímpar (apenas para inteiros)
    if num == int(num):
      if num % 2 == 0:
        print(f"O número {int(num)} é Par.")
      else:
        print(f"O número {int(num)} é Ímpar.")
    else:
      print("Números decimais não são considerados pares ou ímpares.")

  except ValueError:
    print("Entrada inválida. Por favor, digite um número válido.")

def converter_texto_maiusculo():
  """Converte um texto digitado pelo usuário para maiúsculas."""
  texto = input("Digite um texto para converter para maiúsculas: ")
  print(f"Texto em maiúsculas: {texto.upper()}")

def menu():
  """Exibe o menu de opções e executa a função escolhida."""
  while True:
    print("\n--- Menu de Operações --- ")
    print("1. Lançar Dado")
    print("2. Gerar e Analisar 5 Números Aleatórios")
    print("3. Verificar Tipo e Paridade de um Número")
    print("4. Converter Texto para Maiúsculas")
    print("5. Sair")

    escolha = input("Escolha uma opção (1-5): ")

    if escolha == '1':
      lancar_dado()
    elif escolha == '2':
      gerar_e_analisar_numeros()
    elif escolha == '3':
      verificar_numero_completo()
    elif escolha == '4':
      converter_texto_maiusculo()
    elif escolha == '5':
      print("Saindo do programa. Até mais!")
      break
    else:
      print("Opção inválida. Por favor, escolha um número entre 1 e 5.")

# Chama a função do menu para iniciar o programa
menu()
```

```python
# @title
import random
import threading
import queue

# Função para obter entrada com tempo limite
def input_with_timeout(prompt, timeout):
    q = queue.Queue()

    def get_input():
        try:
            q.put(input(prompt))
        except EOFError: # Handles when input stream is closed unexpectedly
            q.put(None)

    thread = threading.Thread(target=get_input)
    thread.daemon = True
    thread.start()
    try:
        return q.get(timeout=timeout)
    except queue.Empty:
        return None

all_questions_by_difficulty = {
    "facil": {
        "A inteligência artificial pode aprender com dados? (sim/não) ": "sim",
        "A internet das coisas (IoT) conecta objetos físicos à internet? (sim/não) ": "sim",
        "5G é uma tecnologia de rede sem fio de quinta geração? (sim/não) ": "sim",
        "Python é uma linguagem de programação? (sim/não) ": "sim",
        "Um SSD é mais rápido que um HDD? (sim/não) ": "sim",
        "Os drones são veículos aéreos não tripulados? (sim/não) ": "sim",
        "A cibersegurança protege sistemas de ataques digitais? (sim/não) ": "sim",
        "Um byte é composto por 8 bits? (sim/não) ": "sim",
        "O mouse é um dispositivo de entrada? (sim/não) ": "sim",
        "A impressora é um dispositivo de saída? (sim/não) ": "sim"
    },
    "medio": {
        "Blockchain é a tecnologia por trás do Bitcoin? (sim/não) ": "sim",
        "Realidade Virtual (RV) é a mesma coisa que Realidade Aumentada (RA)? (sim/não) ": "não",
        "A computação em nuvem permite armazenar dados em servidores remotos? (sim/não) ": "sim",
        "JavaScript é uma linguagem de programação usada para front-end e back-end? (sim/não) ": "sim",
        "Machine Learning é um subcampo da inteligência artificial? (sim/não) ": "sim",
        "O HTML é uma linguagem de programação? (sim/não) ": "não",
        "A Realidade Aumentada sobrepõe informações digitais ao mundo real? (sim/não) ": "sim",
        "Um firewall protege redes contra acessos não autorizados? (sim/não) ": "sim",
        "SQL é uma linguagem de programação para gerenciar bancos de dados? (sim/não) ": "sim",
        "A realidade virtual exige óculos especiais? (sim/não) ": "sim",
        "A VPN (Rede Privada Virtual) torna sua conexão 100% anônima? (sim/não) ": "não"
    },
    "dificil": {
        "Os chatbots são programas de computador que simulam conversas humanas? (sim/não) ": "sim",
        "O sistema operacional é o hardware principal de um computador? (sim/não) ": "não",
        "Um ataque de DDoS (Distributed Denial of Service) sobrecarrega um servidor com tráfego? (sim/não) ": "sim",
        "A criptografia de ponta a ponta garante que apenas o remetente e o destinatário podem ler a mensagem? (sim/não) ": "sim",
        "Python é uma linguagem compilada? (sim/não) ": "não",
        "O IPv6 substituiu completamente o IPv4 em toda a internet? (sim/não) ": "não",
        "A Lei de Moore afirma que o número de transistores em um microchip dobra a cada dois anos? (sim/não) ": "sim",
        "Um algoritmo de 'força bruta' é sempre a forma mais eficiente de resolver um problema? (sim/não) ": "não"
    }
}

print("Bem-vindo ao Quiz de Tecnologia!\nResponda 'sim' ou 'não' para cada pergunta.\nVocê terá 10 segundos para responder a cada uma.")

while True:
    score = 0
    incorrect_answers = 0

    while True:
        difficulty = input("\nEscolha o nível de dificuldade (facil, medio, dificil): ").lower()
        if difficulty in all_questions_by_difficulty:
            questions_and_answers = all_questions_by_difficulty[difficulty]
            break
        else:
            print("Nível de dificuldade inválido. Por favor, escolha entre 'facil', 'medio' ou 'dificil'.")

    while True:
        try:
            num_questions_input = int(input(f"Quantas perguntas você gostaria de responder no nível '{difficulty}' (máx. {len(questions_and_answers)})? "))
            if 0 < num_questions_input <= len(questions_and_answers):
                num_questions_to_ask = num_questions_input
                break
            else:
                print(f"Por favor, insira um número entre 1 e {len(questions_and_answers)}.")
        except ValueError:
            print("Entrada inválida. Por favor, insira um número inteiro.")

    print(f"\nComeçando o jogo com {num_questions_to_ask} perguntas no nível {difficulty}... Você terá 10 segundos para responder a cada pergunta.")

    # Garante que as perguntas não se repitam na mesma rodada
    questions_to_ask_list = random.sample(list(questions_and_answers.keys()), num_questions_to_ask)

    for i, question_chosen in enumerate(questions_to_ask_list):
        correct_answer = questions_and_answers[question_chosen]
        print(f"\n--- Pergunta {i+1} de {num_questions_to_ask} ---")

        user_answer_raw = input_with_timeout(question_chosen, 10) # 10 segundos de tempo limite

        if user_answer_raw is None:
            print("⏳ Tempo esgotado!")
            user_answer = ""
        else:
            user_answer = user_answer_raw.lower()

        if user_answer == correct_answer:
            print(f"✅ Correto! A resposta era '{correct_answer}'. Muito bem!")
            score += 1
        elif user_answer in ['sim', 'não']: # Only count as incorrect if a valid 'sim' or 'não' was given but wrong
            print(f"❌ Incorreto! A resposta correta era '{correct_answer}'. Tente de novo!")
            incorrect_answers += 1
        else: # Invalid input or timed out
            print("⚠️ Resposta inválida ou tempo esgotado! A resposta correta era '" + correct_answer + "'.")
            incorrect_answers += 1
        print("-------------------------------------")

    print(f"\n✨ Fim da Rodada! ✨")
    print(f"Você acertou: {score} perguntas! ✅")
    print(f"Você errou: {incorrect_answers} perguntas! ❌")
    print(f"Sua pontuação final nesta rodada: {score} de {num_questions_to_ask}.")

    play_again = input("\nDeseja jogar novamente? (sim/não): ").lower()
    if play_again != 'sim':
        print("\nObrigado por jogar! Até a próxima! 👋")
        break
```
